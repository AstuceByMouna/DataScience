## 1. Domain Analysis


 <b><font color="green">Dataset Overview:</b>

* The provided dataset contains information related to skin disease diagnosis.
    
* It includes various clinical features as well as an age column.
    
* The target variable, 'class,' represents the disease classification.
    
<b><font color="green">Domain Understanding:</b>

* Skin diseases often exhibit diverse symptoms and characteristics.
    
* Accurate diagnosis is crucial for effective treatment.
    
* Machine learning can assist dermatologists in diagnosing skin conditions based on clinical features.
    
<b><font color="green">Dataset Characteristics:</b>

* Features include both numerical (integer) and categorical data.
    
* 'Age' is a relevant feature, but it needs to be converted to a numerical format.
    
* 'Class' is the target variable with multiple possible disease classifications.
    

* Attributes:
    The dataset contains a total of 34 attributes. Of these, 33 are linear-valued (likely numerical) attributes, and one is nominal (likely categorical). But , all the data are encoded and are in integer format except for age, which is in object format.
    

* Diseases:
    The dataset focuses on the differential diagnosis of erythemato-squamous diseases, which include psoriasis, seborrheic dermatitis, lichen planus, pityriasis rosea, chronic dermatitis, and pityriasis rubra pilaris.

    
* Clinical Evaluation: 
    Initially, patients were clinically evaluated using 12 features. These features may represent various clinical characteristics or symptoms associated with the diseases.
    

* Histopathological Evaluation:
     Skin samples were taken from the patients to evaluate 22 histopathological features. These features are likely related to the microscopic examination of skin tissue samples.

    
* Family History: 
     There is a feature indicating family history, which has a binary value: 1 if any of these diseases have been observed in the family, and 0 otherwise.

    
* Age:
    The dataset includes an age feature representing the age of the patients.

    
* Degree of Features: 
     For most of the features (both clinical and histopathological), values range from 0 to 3, where 0 indicates the feature was not present, 3 indicates the largest amount possible, and 1 or 2 indicate intermediate values. This suggests a degree of severity or presence for each feature.
    
    
* Patient Information: 
    The dataset does not include patient names or ID numbers, likely for privacy reasons.

    
  
    
    
### <b><font color="green">Features in the Dataset</b>
##### Clinical Features:
1. **Erythema:** Redness of the skin, often indicating inflammation or irritation.
2. **Scaling:** Presence of scales or flaking of the skin, associated with conditions like psoriasis or dermatitis.
3. **Definite Borders:** Indicates well-defined boundaries of skin lesions.
4. **Itching:** Common symptom of skin disorders.
5. **Koebner Phenomenon:** Development of skin lesions at sites of trauma or injury.
6. **Polygonal Papules:** Small, raised skin lesions with polygonal shape.
7. **Follicular Papules:** Papules involving hair follicles.
8. **Oral Mucosal Involvement:** Involvement of the oral mucosa.
9. **Knee and Elbow Involvement:** Distribution of skin lesions on joints.
10. **Scalp Involvement:** Presence of skin issues on the scalp.

##### Historical Features:
11. **Family History:** Family history of skin disorders.
12. **Melanin Incontinence:** Leakage of melanin pigment from melanocytes.
13. **Eosinophils in the Infiltrate:** Presence of eosinophils in the inflammatory infiltrate.
14. **PNL Infiltrate:** Presence of polymorphonuclear leukocytes in the infiltrate.
15. **Fibrosis of the Papillary Dermis:** Excess fibrous tissue formation in the dermal layer.
16. **Exocytosis:** Migration of lymphocytes into the epidermis.
17. **Acanthosis:** Thickening of the epidermis.
18. **Hyperkeratosis:** Thickening of the stratum corneum.
19. **Parakeratosis:** Retention of cell nuclei in the stratum corneum.
20. **Clubbing of the Rete Ridges:** Changes in rete ridges.
21. **Elongation of the Rete Ridges:** Elongation of rete ridges.
22. **Thinning of the Suprapapillary Epidermis:** Thinning of this layer.
23. **Spongiform Pustule:** A type of skin lesion.
24. **Munro Microabscess:** Collections of neutrophils in the stratum corneum.
25. **Focal Hypergranulosis:** Increase in granular layer thickness.
26. **Disappearance of the Granular Layer:** Changes in the granular layer.
27. **Vacuolization and Damage of Basal Layer:** Formation of vacuoles in cells.
28. **Spongiosis:** Accumulation of fluid between epidermal cells.
29. **Saw-Tooth Appearance of Rete's:** Structural changes in rete ridges.
30. **Follicular Horn Plug:** Keratinous material in hair follicles.

##### Demographic and Target Variables:
31. **Perifollicular Parakeratosis:** Parakeratosis around hair follicles.
32. **Inflammatory Mononuclear Infiltrate:** Presence of mononuclear immune cells.
33. **Band-Like Infiltrate:** Inflammatory infiltrate forming a band-like pattern.
34. **Age:** Age of the patients.
35. **Class (Target Variable):** Specific skin disorder or disease being diagnosed.

---

This dataset contain a diverse set of clinical, histopathological, and demographic features related to skin disorders. It can be used for diagnostic purposes and predictive modeling to classify different skin conditions based on the provided features.




### <b><font color="green">Dependent Variable(y):</b>
    
  * An integer column indicating the class or target variable (i.e., disease classification) with integer values.
    
  * It has a total of 6 classes
    
  * Since the data set is encoded, class contains 6 types of skin diseases. Our job is to classify the erythemato-squamous    disease in one of the following types:

1. psoriasis 

2. seboreic dermatitis

3. lichen planus 

4. pityriasis rosea

5. cronic dermatitis

6. pityriasis rubra pilaris


    
<b><font color="green">Business Goal:</b>
    
It is a muti class  classification problem.
    
Our goal is to classify disorders into groups depending on its features. We are going to create a Machine Learning Model to classify the disorder in to 1 of the 6 categories. This model can then be used to help the doctors to make informed and quick decisions with their patients' ailments.  
    
<b><font color="green">Benifits of the model:</b>
    
•	Improved accuracy and efficiency of diagnosis: 
    
    The Machine learning model  created from this dataset can be trained on large datasets of skin images and clinical data to learn to identify the subtle features that distinguish between different types of skin disorders. This can lead to more accurate and efficient diagnosis, especially for rare or difficult-to-diagnose disorders.

•	Reduced reliance on expert clinicians: 
    
    The machine learning model created can be used to provide preliminary diagnoses, which can then be reviewed and confirmed by expert clinicians. This can help to reduce the workload on clinicians and make it easier for patients to get access to early care.

•	Early detection and intervention:
    
    The model created can be used to identify skin conditions when they are most treatable. This can lead to better patient outcomes and reduced healthcare costs.

•	Personalized treatment recommendations: 
    
    The machine learning model created can be used to analyze a patient's individual medical history and skin lesion characteristics to develop personalized treatment recommendations. This can help to improve the effectiveness and efficiency of treatment.

```python
# class distribution bar graph 
sns.countplot(x='class', data=data)
plt.show()
```

```python
selected_univariate_features = ["erythema", "scaling", "definite_borders", "itching", "koebner_phenomenon"]
```

```python
plt.figure(figsize=(15, 6))
for i, feature in enumerate(selected_univariate_features, 1):
    plt.subplot(2, 5, i)
    sns.countplot(data=data, x=feature)
    plt.title(f"Univariate Analysis of {feature}")
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Bivariate Analysis

```python
selected_bivariate_features = [("erythema", "scaling"), ("itching", "koebner_phenomenon"),
                                ("family_history", "melanin_incontinence"),
                               ("thinning_of_the_suprapapillary_epidermis", "spongiform_pustule")]
```

```python
plt.figure(figsize=(15, 9))
for i, (feature1, feature2) in enumerate(selected_bivariate_features, 1):
    plt.subplot(2, 3, i)
    sns.barplot(data=data, x=feature1, y=feature2, hue="class", palette="viridis")
    plt.title(f"Bivariate Analysis of {feature1} vs. {feature2}")
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation between 'family_history' and other features
correlations = data.corr()["family_history"].drop("family_history")
```

```python
# Sort the features by their correlation with 'family_history'
sorted_correlations = correlations.abs().sort_values(ascending=False)
```

```python
# Plot the correlation values
plt.figure(figsize=(12, 6))
sns.barplot(x=sorted_correlations.index, y=sorted_correlations.values, palette="viridis")
plt.xticks(rotation=90)
plt.title("Correlation with Family History")
plt.xlabel("Features")
plt.ylabel("Absolute Correlation")
plt.show()
```

```python
# Print the feature with the highest impact
most_impacted_feature = sorted_correlations.index[0]
print(f"The feature mostly impacted by Family History is '{most_impacted_feature}' with a correlation of {sorted_correlations.iloc[0]:.2f}")
```

```python
# from the above correlation plotting a graph b/w family_history and knee_and_elbow_involvement
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='family_history', hue='knee_and_elbow_involvement', palette='Set2')
plt.title("Bivariate Analysis: Family History vs. Knee and Elbow Involvement")
plt.xlabel("Family History")
plt.ylabel("Count")
plt.legend(title='Knee and Elbow Involvement', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
# bivariate analysis graph between 'itching' and 'scaling'
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='itching', y='scaling', palette='Set2')
plt.title("Bivariate Analysis: Itching vs. Scaling")
plt.xlabel("Itching")
plt.ylabel("Scaling")
plt.show()
```

```python
selected_features = ['exocytosis', 'acanthosis', 'hyperkeratosis', 'parakeratosis']
# pairplot for bivariate analysis
sns.set(style="ticks")
sns.pairplot(data=data, vars=selected_features, diag_kind='kde')
plt.suptitle("Bivariate Analysis Pairplot for exocytosis, acanthosis, hyperkeratosis and parakeratosis ", y=1.02)
plt.show()
```

```python
# bivariate analysis using bubble chart between oral_mucosal_involvement,knee_and_elbow_involvement and scalp_involvement  

feature1 = 'oral_mucosal_involvement'
feature2 = 'knee_and_elbow_involvement'
feature3 = 'scalp_involvement'

plt.figure(figsize=(10, 8))
sns.scatterplot(data=data, x=feature1, y=feature2, size=feature3, sizes=(50, 500), hue=feature3, palette='viridis')
plt.title("Bubble Chart of oral_mucosal_involvement,knee_and_elbow_involvement and scalp_involvement")
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.legend(title=feature3, loc='upper right')
plt.show()
```
